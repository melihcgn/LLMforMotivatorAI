```python
import pandas as pd
import re

# Load the CSV file into a DataFrame
df = pd.read_csv('motivational_quotes.csv')
```

```python
def split_quote_author(quote_author):
    # Match the quote and author with `/` or `—`
    match = re.match(r'^(.*?)(?:\s*[-/]\s*(.*))$', quote_author.strip())

    if match:
        quote = match.group(1).strip()
        author = match.group(2).strip()
        print(author)
    else:
        # If no match, assign "Unknown" as author
        quote = quote_author.strip()
        author = 'Unknown'

    return quote, author
```

```python
# Function to clean the quote
def clean_quote(quote):
    # Remove extra spaces and unwanted quotation marks
    quote = re.sub(r'["“”]', '', quote)
    quote = re.sub(r'\s+', ' ', quote).strip()
    return quote
```

```python

# Function to clean the author
def clean_author(author):
    # Remove extra spaces and capitalize names
    author = re.sub(r'\s+', ' ', author).strip()
    author = author.title()  # Ensure consistent capitalization
    return author
```

```python
# Apply cleaning functions to the 'quote' column (assuming both quote and author are in this column)
df[['quote', 'author']] = df['quote'].apply(lambda x: pd.Series(split_quote_author(x)))

# Apply the cleaning functions
df['quote'] = df['quote'].apply(clean_quote)
df['author'] = df['author'].apply(clean_author)
```

```python
# Check for duplicates in the quotes
duplicates = df[df.duplicated(subset=['quote'], keep=False)]
if not duplicates.empty:
    print("Duplicate quotes found:")
    print(duplicates)

# Check for missing authors
missing_authors = df[df['author'] == 'Unknown']
if not missing_authors.empty:
    print("Quotes with missing authors:")
    print(missing_authors)
```

```python
# Save the final cleaned dataset to a new CSV file
df.to_csv('final_cleaned_quotes.csv', index=False)
print("Final dataset saved successfully!")
```
